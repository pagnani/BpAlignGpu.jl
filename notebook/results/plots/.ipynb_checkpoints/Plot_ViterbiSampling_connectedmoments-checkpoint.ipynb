```julia
using DelimitedFiles
```

```julia
data = readdlm("store_Cij_ab.txt");
Cseed = data[1,:];
Csamp = data[2,:];
CGS = data[3,:];
```

```julia
data
```

```julia
using KernelDensity
```

```julia
dens_c=kde((Cseed,CGS));
ik = InterpKDE(dens_c);
vect_color = [pdf(ik, Cseed[k], CGS[k]) for k=1:length(Cseed)];
```

```julia
dens_cc=kde((Cseed,Csamp));
ikk = InterpKDE(dens_cc);
vect_color_2 = [pdf(ikk, Cseed[k], Csamp[k]) for k=1:length(Cseed)];
```

```julia
using PyPlot
```

```julia
p = scatter(Cseed, CGS, c=vect_color,s=3, edgecolor="")
```

```julia
p2 = scatter(Cseed, Csamp, c=vect_color_2,s=3, edgecolor="")
```
