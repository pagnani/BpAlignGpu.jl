# 1. epsilon-coupling 

# Plot $\Delta E$ and Hamming distance of the ground states

```julia
path = "../data_epsilon_coupling/"
```

### data obtained with MF

```julia
namefile = path * "run_epsilon_PF00397_mf_100b.txt"
data_epsilon = readdlm(namefile);
param_epsilon = data_epsilon[1,:]

neps = sum(typeof.(data_epsilon[2,:]) .== Float64)
res_epsilon = data_epsilon[2:end,1:neps];
epsilons = res_epsilon[1,:]

μΔEs = res_epsilon[2,:]
σΔEs = res_epsilon[3,:]
μdist = res_epsilon[6,:]
σdist = res_epsilon[7,:];
```

### data obtained with SCE

```julia
namefile = path * "run_epsilon_PF00397_sce_100b_betafixed.txt"
data_epsilon = readdlm(namefile);
param_epsilon = data_epsilon[1,:]

neps = sum(typeof.(data_epsilon[2,:]) .== Float64)
res_epsilon = data_epsilon[2:end,1:neps];

epsilons_sce = res_epsilon[1,:]
μΔEs_sce = res_epsilon[2,:]
σΔEs_sce = res_epsilon[3,:]
μdist_sce = res_epsilon[6,:]
σdist_sce = res_epsilon[7,:];
```

```julia
namefile = path * "run_epsilon_PF00397_sce_100b_betafixed_damp0p5.txt"
data_epsilon = readdlm(namefile);
param_epsilon = data_epsilon[1,:]

neps = sum(typeof.(data_epsilon[2,:]) .== Float64)
res_epsilon = data_epsilon[2:end,1:neps];

epsilons_sce_d = res_epsilon[1,:]
μΔEs_sce_d = res_epsilon[2,:]
σΔEs_sce_d = res_epsilon[3,:]
μdist_sce_d = res_epsilon[6,:]
σdist_sce_d = res_epsilon[7,:];
```

```julia
L=31
```

```julia
pΔ = scatter(-epsilons_sce[1:5], μΔEs_sce[1:5]./L,  yerr=σΔEs_sce[1:5]./L, label="", legend=:topleft, color="red")
scatter!(-epsilons_sce_d, μΔEs_sce_d./L,  yerr=σΔEs_sce_d./L, legend=:topleft, color="red", label="")
#scatter!(-epsilons, μΔEs./L, yerr=σΔEs./L, label="mf", legend=:topleft, color="black")

xlabel!("\$ \\epsilon\$")
ylabel!("\$ \\Delta e_{\\min}\$")
hline!([0.0], label="", color="black")

pdist = scatter(-epsilons_sce[1:5], μdist_sce[1:5], yerr=σdist_sce, label="", legend=:topleft, color="red")
scatter!(-epsilons_sce_d, μdist_sce_d, yerr=σdist_sce_d, legend=:topleft, color="red", label="")
#scatter!(-epsilons, μdist, yerr=σdist, label="mf", legend=:topleft, color="black")

hline!([0.0], label="", color="black")
xlabel!("\$ \\epsilon\$")
ylabel!("Hamming distance")
plot(pΔ, pdist, size=(1000,500), margin=6*Plots.mm)
```

```julia
savefig("epsilon_PF00397.pdf")
```

# Plot at fixed $\beta$ of thermodynamic quantities

### data obtained with SCE

```julia
namefile = path * "run_epsilon_PF00397_sce_100b_betafixed.txt"
data_epsilon = readdlm(namefile);
param_epsilon = data_epsilon[1,:]

neps = sum(typeof.(data_epsilon[2,:]) .== Float64)
res_epsilon = data_epsilon[2:end,1:neps];

epsilons = res_epsilon[1,:]
μE_bet = res_epsilon[8,:]
σE_bet = res_epsilon[9,:]
μU_bet = res_epsilon[10,:]
σU_bet = res_epsilon[11,:]
μS_bet = res_epsilon[12,:]
σS_bet = res_epsilon[13,:];
μpolars = res_epsilon[14,:]
σpolars = res_epsilon[15,:];
```

```julia
namefile = path * "run_epsilon_PF00397_sce_100b_betafixed_damp0p5.txt"
data_epsilon = readdlm(namefile);
param_epsilon = data_epsilon[1,:]

neps = sum(typeof.(data_epsilon[2,:]) .== Float64)
res_epsilon = data_epsilon[2:end,1:neps];

epsilons_d = res_epsilon[1,:]
μE_bet_d = res_epsilon[8,:]
σE_bet_d = res_epsilon[9,:]
μU_bet_d = res_epsilon[10,:]
σU_bet_d = res_epsilon[11,:]
μS_bet_d = res_epsilon[12,:]
σS_bet_d = res_epsilon[13,:];
μpolars_d = res_epsilon[14,:]
σpolars_d = res_epsilon[15,:];
```

```julia
pu = scatter(-epsilons[1:5], μE_bet[1:5]./L, yerr=σE_bet[1:5]./L, label="ground state", legend=:topleft, color = "blue")
scatter!(-epsilons_d, μE_bet_d./L, yerr=σE_bet_d./L, color="blue", label="")
scatter!(-epsilons[1:5], μU_bet[1:5]./L, yerr=σU_bet[1:5]./L, label="average", color = "orange")
scatter!(-epsilons_d, μU_bet_d./L, yerr=σU_bet_d./L, color="orange", label="")

xlabel!("\$ \\epsilon\$")
ylabel!("energy density")
ps = scatter(-epsilons[1:5], μS_bet[1:5]./L, yerr=σS_bet[1:5]./L, legend=:topleft, color = "green", label="")
scatter!(-epsilons_d, μS_bet_d./L, yerr=σS_bet_d./L, color="green", label="")
#scatter!(-epsilons, μpolars, yerr=σpolars, legend=:topleft, color = "black", label="")


xlabel!("\$ \\epsilon\$")
ylabel!("entropy density")
hline!([0.0], label="", color="black", legend=:bottomright)
plot(pu, ps, size=(1000,500), margin=6*Plots.mm)
```

```julia
savefig("epsilon_PF00397_thqs.pdf")
```
